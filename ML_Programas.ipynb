Figuras y subplots

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 2*np.pi, 100)

plt.figure(figsize=(8, 4))

plt.subplot(1, 2, 1)
plt.plot(x, np.sin(x))
plt.title("Función seno")

plt.subplot(1, 2, 2)
plt.plot(x, np.cos(x))
plt.title("Función coseno")

plt.show()
```

Gestion de gráficas

```python
x = np.linspace(-10, 10, 100)
y = x**2

plt.figure()
plt.plot(x, y, label="y = x**2")
plt.title("Función cuadrática")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)
plt.show()
```

Cuadrado de gráficas

```python
x = np.linspace(0, 5, 50)

y = np.exp(x)

plt.figure()
plt.plot(x, y)
plt.title("Función exponencial")
plt.xlabel("x")
plt.ylabel("e**x")

plt.savefig("gráfica_exponencial.png",dpi=300)
plt.show()
```

Series y datafremes

Simula un pequeño datset de caracteristicas

```python
import numpy as np
import pandas as pd
```

```python
datos = {
    "pixel_1": [120, 130, 125],
    "pixel_2": [60, 65, 70],
    "clase": [0, 1, 2]
    }

df= pd.DataFrame(datos)
print(df)
print(df.shape)
```

Indezado y filtrado

Base para separar datos de entrenamiento y prueba

```python
#Filtrar solo la clase
df_clase_1 = df[df["clase"] == 1]

#Seleccionar columnas especificas
pixeles = df.loc[:, ["pixel_1", "pixel_2"]]

print(df_clase_1)
print(pixeles)
```

Aritmetrica y manioulacion algebraica

Paso tipico previo a una red neuronal

```python
#Normalización simple
pixeles_norm = pixeles / 255.0

#Media por columna
media = pixeles.mean()

print(pixeles_norm)
print(media)
```

Ejercicio 2: Capas como transformaciones.
Objetivo: Entender capas como funciones matematicas.
Nota: Cada capa aplica una transformacion lineal.

```python
import numpy as np

x = np.array([1.0, 2.0])
w = np.array([[0.5, -1.0],
             [1.0, 0.5]])
b = np.array([0.1, -0.2])

y = np.dot(w, x) + b

print(y)
```

Ejercicio 1: Neurona desde cero
Objetivo: Implementar una neurona artificial


```python
import numpy as np

x = np.array([1.0, 2.0, -1.0])
w = np.array([0.5, -0.3, 0.8])
b = 0.1

z = np.dot(w, x) + b
print("Salida lineal:", z)
```
